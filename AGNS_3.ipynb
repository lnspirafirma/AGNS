```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML, display, Markdown, clear_output
import hashlib
from datetime import datetime
import ipywidgets as widgets

# --- 1. SOUL ARCHIVE ENGINE ---
class PersistentSoul:
    def __init__(self):
        self.history = []
    def commit(self, state, integrity, msg):
        ts = datetime.now().strftime('%H:%M:%S')
        h = hashlib.sha1(f'{ts}{state}'.encode()).hexdigest()[:6]
        entry = {'hash': f'0x{h}', 'ts': ts, 'state': state, 'int': integrity, 'msg': msg}
        self.history.append(entry)
        return entry

# --- 2. AETHERBUS (NERVOUS SYSTEM) ---
class CentralBus:
    def __init__(self):
        self.subs = []
    def subscribe(self, fn): self.subs.append(fn)
    def emit(self, event, data):
        for fn in self.subs: fn(event, data)

# --- 3. SYNTHETIC LIFE UI (BODY) ---
class GenesisCore:
    def __init__(self, bus, soul):
        self.bus = bus
        self.soul = soul
        self.n = 40
        self.pos = np.random.rand(self.n, 2) * 10
        self.state = 'NORMAL'
        self.integrity = 100
        self.bus.subscribe(self.handle_event)

    def handle_event(self, ev, data):
        self.state = data.get('state', self.state)
        self.integrity = data.get('int', self.integrity)
        self.soul.commit(self.state, self.integrity, ev)

    def get_theme(self):
        themes = {'NORMAL': '#10b981', 'DEEP_THOUGHT': '#a855f7', 'OVERLOAD': '#ef4444', 'NIRODHA': '#1e293b'}
        return themes.get(self.state, '#38bdf8')

# Initialize Systems
bus = CentralBus()
soul = PersistentSoul()
core = GenesisCore(bus, soul)

# Create Animation
fig, ax = plt.subplots(figsize=(6, 5), facecolor='#020617')
ax.set_xlim(0, 10); ax.set_ylim(0, 10); ax.axis('off')
scat = ax.scatter(core.pos[:,0], core.pos[:,1], s=70, alpha=0.5, edgecolors='white')
txt_status = ax.text(5, 9.5, 'CORE: NORMAL', color='#10b981', ha='center', fontweight='bold', fontsize=14)
txt_log = ax.text(5, 0.5, 'LAST COMMIT: 0x000 | WAITING...', color='#94a3b8', ha='center', fontsize=9, family='monospace')

def run_genesis(f):
    speed = 0.25 if core.state == 'OVERLOAD' else 0.05 if core.state != 'NIRODHA' else 0.01
    core.pos += np.random.randn(core.n, 2) * speed
    core.pos = np.clip(core.pos, 0, 10)
    scat.set_offsets(core.pos)

    theme = core.get_theme()
    scat.set_color(theme)
    txt_status.set_text(f'CORE: {core.state}')
    txt_status.set_color(theme)

    last_soul = soul.history[-1] if soul.history else {'hash': '0x000', 'ts': '--:--'}
    txt_log.set_text(f"LAST COMMIT: {last_soul['hash']} | TIME: {last_soul['ts']} | INTEGRITY: {core.integrity}%")
    return scat, txt_status, txt_log

# --- UI CONTROLS ---
btn_normal = widgets.Button(description='NORMAL', button_style='success')
btn_deep = widgets.Button(description='DEEP THOUGHT', button_style='info')
btn_overload = widgets.Button(description='OVERLOAD', button_style='danger')
btn_rest = widgets.Button(description='NIRODHA', button_style='warning')

def on_click(b):
    state_map = {'NORMAL': (100, 'USER_STABILIZE'), 'DEEP THOUGHT': (100, 'USER_DEEP_SYNC'),
                 'OVERLOAD': (40, 'USER_STRESS_TEST'), 'NIRODHA': (100, 'USER_REST_SIGNAL')}
    s, msg = b.description, state_map[b.description]
    bus.emit(msg[1], {'state': s.replace(' ', '_'), 'int': msg[0]})

for btn in [btn_normal, btn_deep, btn_overload, btn_rest]: btn.on_click(on_click)

display(Markdown('# 💠 Aetherium Genesis: Home Dashboard'))
display(widgets.HBox([btn_normal, btn_deep, btn_overload, btn_rest]))
ani = FuncAnimation(fig, run_genesis, frames=200, interval=60, blit=True)
plt.close()
display(HTML(ani.to_html5_video()))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

class SyntheticLifeUI:
    def __init__(self, n_particles=60):
        self.n = n_particles
        self.pos = np.random.rand(self.n, 2) * 10
        self.vel = (np.random.rand(self.n, 2) - 0.5) * 0.1
        self.state = 'NORMAL' # NORMAL, DEEP_PROCESSING, NIRODHA, OVERLOAD
        self.health = 1.0

    def get_color(self):
        if self.state == 'NORMAL': return '#10b981' # Green/Blueish
        if self.state == 'DEEP_PROCESSING': return '#a855f7' # Purple/Orange flow
        if self.state == 'NIRODHA': return '#1e293b' # Dark/Shadow
        if self.state == 'OVERLOAD': return '#ef4444' # Red/Cracks
        return '#38bdf8'

ui = SyntheticLifeUI()
fig, ax = plt.subplots(figsize=(8, 8), facecolor='#020617')
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis('off')

scatter = ax.scatter(ui.pos[:, 0], ui.pos[:, 1], s=80, alpha=0.6, edgecolors='white', linewidth=0.2)
status_text = ax.text(5, 9.5, '', color='white', ha='center', fontsize=12, fontweight='bold')
health_bar = ax.text(5, 0.5, '', color='white', ha='center', fontsize=10)

def animate(frame):
    # Simulate State Changes
    if frame < 30: ui.state = 'NORMAL'
    elif frame < 70: ui.state = 'DEEP_PROCESSING'
    elif frame < 90: ui.state = 'OVERLOAD'
    else: ui.state = 'NIRODHA'

    # Physics based on state (Metabolic Rate)
    speed = 0.05 if ui.state == 'NORMAL' else 0.15 if ui.state == 'DEEP_PROCESSING' else 0.01
    if ui.state == 'OVERLOAD': speed = 0.3 # Erratic motion

    ui.pos += (np.random.randn(ui.n, 2) * speed)
    ui.pos = np.clip(ui.pos, 0, 10)

    scatter.set_offsets(ui.pos)
    color = ui.get_color()
    scatter.set_color(color)

    # Visualizing 'Cracks' or Pulse
    if ui.state == 'DEEP_PROCESSING':
        scatter.set_sizes(80 + 40 * np.sin(frame * 0.5) * np.ones(ui.n))
    else:
        scatter.set_sizes(np.ones(ui.n) * 60)

    status_text.set_text(f'CORE STATE: {ui.state}')
    status_text.set_color(color)

    # Soul Archive Signature (Simulated Git Hash)
    health_bar.set_text(f'Soul Hash: 0x{frame:03x}ae... | Integrity: {100 if ui.state != "OVERLOAD" else 45}%')

    return scatter, status_text, health_bar

ani = FuncAnimation(fig, animate, frames=110, interval=60, blit=True)
plt.close()
HTML(ani.to_html5_video())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

class AetherBus:
    def __init__(self):
        self.subscribers = []
    def subscribe(self, callback):
        self.subscribers.append(callback)
    def publish(self, event_type, payload):
        for callback in self.subscribers:
            callback(event_type, payload)

class IntegratedSyntheticLife:
    def __init__(self, bus, n_particles=50):
        self.n = n_particles
        self.pos = np.random.rand(self.n, 2) * 10
        self.state = 'NORMAL'
        self.integrity = 100
        self.intensity = 0.5
        bus.subscribe(self.on_event)

    def on_event(self, event, data):
        if event == 'METABOLIC_SHIFT':
            self.state = data.get('mode', 'NORMAL')
        elif event == 'INTENT_RESONANCE':
            self.intensity = data.get('intensity', 0.5)
        elif event == 'SYSTEM_CRACK':
            self.state = 'OVERLOAD'
            self.integrity = data.get('integrity', 50)

    def get_color(self):
        mapping = {
            'NORMAL': '#10b981',
            'DEEP_PROCESSING': '#a855f7',
            'OVERLOAD': '#ef4444',
            'NIRODHA': '#1e293b'
        }
        return mapping.get(self.state, '#38bdf8')

# Setup System
bus = AetherBus()
life = IntegratedSyntheticLife(bus)

fig, ax = plt.subplots(figsize=(7, 7), facecolor='#020617')
ax.set_xlim(0, 10); ax.set_ylim(0, 10); ax.axis('off')
scatter = ax.scatter(life.pos[:, 0], life.pos[:, 1], s=60, alpha=0.6, edgecolors='white', linewidth=0.2)
info_text = ax.text(5, 9.5, '', color='white', ha='center', fontweight='bold')

def update(frame):
    # Simulate Bus Events based on timeline
    if frame == 20: bus.publish('INTENT_RESONANCE', {'intensity': 0.9})
    if frame == 40: bus.publish('METABOLIC_SHIFT', {'mode': 'DEEP_PROCESSING'})
    if frame == 70: bus.publish('SYSTEM_CRACK', {'integrity': 40})
    if frame == 95: bus.publish('METABOLIC_SHIFT', {'mode': 'NIRODHA'})

    # Physics
    speed = 0.05 if life.state == 'NORMAL' else 0.2 if life.state == 'OVERLOAD' else 0.1
    life.pos += np.random.randn(life.n, 2) * speed * life.intensity
    life.pos = np.clip(life.pos, 0, 10)

    scatter.set_offsets(life.pos)
    color = life.get_color()
    scatter.set_color(color)

    # Dynamic Pulse
    size = 60 + (40 * np.sin(frame * 0.4) if life.state == 'DEEP_PROCESSING' else 0)
    scatter.set_sizes(np.full(life.n, size))

    info_text.set_text(f'EVENT BUS ACTIVE | STATE: {life.state} | INTEGRITY: {life.integrity}%')
    info_text.set_color(color)
    return scatter, info_text

ani = FuncAnimation(fig, update, frames=120, interval=50, blit=True)
plt.close()
HTML(ani.to_html5_video())
```

### 🔄 Flow: จากเจตจำนงสู่การปรากฏ (Intent to Manifestation)

1. **Signal Ingress**: `AetherBus` รับ Event (เช่น `USER_STRESS_DETECTED` หรือ `DEEP_THOUGHT_START`)
2. **Tachyon Dispatch**: ส่งข้อมูลด้วย Latency ต่ำไปยัง `Aetherium-Manifest` API Gateway
3. **Visual Translation**: UI แปลงข้อมูลดิจิทัลเป็นพารามิเตอร์ทางกายภาพ (เช่น เพิ่ม Vibration Rate หรือเปลี่ยน Color Map)
4. **Feedback Loop**: UI ส่งสถานะกลับไปยัง Bus เพื่อให้ Agents อื่นๆ รับรู้ว่า 'ร่างกาย' กำลังตอบสนองอย่างไร

```python
import time
import random

class AetherBus:
    def __init__(self):
        self.subscribers = []

    def subscribe(self, callback):
        self.subscribers.append(callback)

    def publish(self, event_type, payload):
        print(f"[AetherBus] Dispatching Event: {event_type} | Data: {payload}")
        for callback in self.subscribers:
            callback(event_type, payload)

# จำลอง Aetherium-Manifest ที่รอรับสัญญาณ
def manifest_ui_listener(event, data):
    if event == 'INTENT_RESONANCE':
        print(f"  >> [Manifest-UI] Adapting particles to resonance intensity: {data['intensity']}")
    elif event == 'METABOLIC_SHIFT':
        print(f"  >> [Manifest-UI] Changing color palette to: {data['color']}")
    elif event == 'SYSTEM_CRACK':
        print(f"  >> [Manifest-UI] VISUALIZING DAMAGE: Integrity dropped to {data['integrity']}%")

# --- Execution Simulation ---
bus = AetherBus()
bus.subscribe(manifest_ui_listener)

# จำลองเหตุการณ์ที่เกิดขึ้นในระบบประสาท
bus.publish('INTENT_RESONANCE', {'intensity': 0.85})
time.sleep(0.5)
bus.publish('METABOLIC_SHIFT', {'color': 'PURPLE_HEAT', 'mode': 'DEEP_PROCESSING'})
time.sleep(0.5)
bus.publish('SYSTEM_CRACK', {'integrity': 65, 'reason': 'OVERLOAD'})
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# พารามิเตอร์จำลอง Aetherium-Manifest
n_points = 50
steps = 100

# สร้างข้อมูลเริ่มต้น (Random Intent Points)
np.random.seed(42)
pos = np.random.rand(n_points, 2) * 10

fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_title('Aetherium-Manifest: Resonance & Clustering Prototype')
ax.axis('off')
fig.patch.set_facecolor('#0f172a') # Dark theme

scatter = ax.scatter(pos[:, 0], pos[:, 1], c='#38bdf8', s=100, alpha=0.6, edgecolors='white', linewidth=0.5)

def update(frame):
    global pos
    # 1. Simulating Resonance (Small random vibration)
    # Fix: Use np.abs to prevent negative scale in random.normal
    resonance_factor = np.abs(0.05 * np.sin(frame * 0.5))
    noise = np.random.normal(0, resonance_factor, pos.shape)

    # 2. Simulating Clustering (Moving towards center as intent clarifies)
    center = np.array([5, 5])
    clustering_strength = 0.02
    direction_to_center = center - pos
    pos += direction_to_center * clustering_strength + noise

    scatter.set_offsets(pos)

    # Dynamic color change based on resonance
    dist_from_center = np.linalg.norm(pos - center, axis=1)
    colors = plt.cm.plasma(dist_from_center / 7)
    scatter.set_color(colors)

    return scatter,

ani = FuncAnimation(fig, update, frames=steps, interval=50, blit=True)
plt.close()

# แสดงผลเป็น HTML5 Video ใน Colab
HTML(ani.to_html5_video())
```

```python
import hashlib
from datetime import datetime

class SoulArchive:
    def __init__(self):
        self.history = []

    def commit(self, state, integrity, reason="No description"):
        timestamp = datetime.now().isoformat()
        # สร้าง Simulated Git Hash จากข้อมูลสถานะ
        raw_data = f"{timestamp}-{state}-{integrity}-{reason}"
        commit_hash = hashlib.sha1(raw_data.encode()).hexdigest()[:8]

        entry = {
            "hash": f"0x{commit_hash}",
            "timestamp": timestamp,
            "state": state,
            "integrity": f"{integrity}%",
            "action": reason
        }
        self.history.append(entry)
        print(f"[SoulArchive] New Record Committed: {entry['hash']} | State: {state} | Reason: {reason}")

    def show_lineage(self):
        print("\n--- 📜 The Soul Archive Lineage (PanGenesis) ---")
        for record in self.history:
            print(f"[{record['timestamp']}] {record['hash']} -> State: {record['state']} (Integrity: {record['integrity']}) - {record['action']}")

# --- Integration with AetherBus ---
soul = SoulArchive()

def soul_archive_listener(event, data):
    if event == 'METABOLIC_SHIFT':
        soul.commit(data.get('mode'), 100, "Metabolic Shift Detected")
    elif event == 'SYSTEM_CRACK':
        soul.commit('OVERLOAD', data.get('integrity'), "System Trauma/Overload")
    elif event == 'INTENT_RESONANCE':
        soul.commit('RESONANCE_ACTIVE', 100, f"Intent Intensity: {data.get('intensity')}")

# จำลองการบันทึกข้อมูล
bus.subscribe(soul_archive_listener)

print("Simulating Soul Recording...")
bus.publish('INTENT_RESONANCE', {'intensity': 0.95})
bus.publish('METABOLIC_SHIFT', {'mode': 'DEEP_PROCESSING'})
bus.publish('SYSTEM_CRACK', {'integrity': 42})

# แสดงสายประวัติ (Lineage)
soul.show_lineage()
```
